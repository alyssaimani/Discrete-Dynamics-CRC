# Dynamic Simulations

### Equilibrium point 2

```python
from mathdyn.equations import *
from datatypes.loader import *
th = 300
p, init, ival = get_params("params.yaml", "eq2")
t_val = np.arange(ival.t0, th, ival.h)
```

```python
H_eq, C_eq, B_eq = equilibrium(p.point, p.r_H, p.r_C, p.r_B, p.alpha, p.beta, p.gamma, p.m_1, p.m_2, p.g, p.k, p.k_B, p.b_0)
eigenvalues, re, im = find_jacobian(H_eq, C_eq, B_eq, p.r_H, p.r_C, p.r_B, p.alpha, p.beta, p.gamma, p.m_1, p.m_2, p.g, p.k, p.k_B, p.b_0)
phi = find_phi(eigenvalues, ival.h)
print("equilibrium point:", f"({H_eq}, {C_eq}, {B_eq})")
```

```python
lam1, lam2, lam3 = get_lambda(p.point, p.r_H, p.r_C, p.r_B, p.alpha, p.beta, p.gamma, p.m_1, p.m_2, p.g, p.k, p.k_B, p.b_0)
print("lambda 1:", lam1)
print("lambda 2:", lam2)
print("lambda 3:", lam3)
print("eigenvalues: ", eigenvalues)
```

```python
# NSFD
H_nsfd, C_nsfd, B_nsfd = find_nsfd(init.H_nol, init.C_nol, init.B_nol, 
                                   phi, t_val, p.r_H, p.r_C, p.r_B, 
                                   p.alpha, p.beta, p.gamma, p.m_1, p.m_2, 
                                   p.g, p.k, p.k_B, p.b_0)
# Runge Kutta
rk_H, rk_C, rk_B = find_rk(init.H_nol, init.C_nol, init.B_nol, 
                            ival.t0, th, t_val, 
                            p.r_H, p.r_C, p.r_B, p.alpha, p.beta, p.gamma, 
                            p.m_1, p.m_2, p.g, p.k, p.k_B, p.b_0)
# en
get_endpoint("runge kutta", rk_H, rk_C, rk_B, th)
get_endpoint("nsfd",H_nsfd, C_nsfd, B_nsfd, th)
```

```python
diff_H, diff_C, diff_B = find_diff(H_nsfd, C_nsfd, B_nsfd, 
                                   rk_H, rk_C, rk_B, t_val)
max_H = max(diff_H)
max_C = max(diff_C)
max_B = max(diff_B)
print("max diff H, C, B", [max_H, max_C, max_B])
plot_all(t_val, H_nsfd, C_nsfd, B_nsfd, rk_H, rk_C, rk_B, diff_H, diff_C, diff_B)
```
